```python
#Funciton to replace all nan value by mode oof their common group
def replace_nan_with_mode(group):
    mode=group.mode().iloc[0]
    return group.fillna(mode)
```

```python
# Group features in dataset using Customer ID as common 
grouped_df=df.groupby("Customer_ID")
```

```python
# Replace all misssing categorical feature by mode of the feature grouped using Customer_ID
df[c_features]=grouped_df[c_features].apply(replace_nan_with_mode)
```

### *Convert temporal feature Credit_History_Age in terms of month*

```python
df["Credit_History_Age"].head(20)
```

```python
import re
```

```python
# A funciton to convert to get number or years and months from string
def get_years(x):
    years=re.search("(\d+) Years",x)
    return int(years.group(1))
def get_months(x):
    years=re.search("(\d+) Months",x)
    return int(years.group(1)) 
```

```python
## Check if it works correctly
get_years(df["Credit_History_Age"][0])*12+get_months(df["Credit_History_Age"][0])
```

```python
df["Credit_History_Age"]=df["Credit_History_Age"].apply(lambda x:x if type(x)==float else get_years(x)*12+get_months(x))
```

```python
df["Credit_History_Age"].head(20)
```

```python
def replace_nan_with_median(group):
    median=group.median()
    return group.fillna(median)
```

```python
grouped_df=df.groupby("Customer_ID")
```

```python
df["Credit_History_Age"]=grouped_df["Credit_History_Age"].apply(replace_nan_with_median)
```

```python
print(df["Credit_History_Age"].describe())
```
